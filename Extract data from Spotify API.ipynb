```python
# Get URL of top 200 song to extract song feature on spotify
URL = top200['Link'].str.split("https://open.spotify.com/track/", n = 1, expand = True)
track_ids = URL[1].values.tolist()
print(len(track_ids))
print(track_ids)
```

```python
# function to get song features
def getTrackFeatures(id):
    features_info = sp.audio_features(id)
    features_track = sp.track(id)

    # Track features
    
    popularity = features_track['popularity']
    id = features_info[0]['id']
    acousticness = features_info[0]['acousticness']
    danceability = features_info[0]['danceability']
    energy = features_info[0]['energy']
    key = features_info[0]['key']
    mode = features_info[0]['mode']
    instrumentalness = features_info[0]['instrumentalness']
    liveness = features_info[0]['liveness']
    loudness = features_info[0]['loudness']
    speechiness = features_info[0]['speechiness']
    valence = features_info[0]['valence']
    tempo = features_info[0]['tempo']
    time_signature = features_info[0]['time_signature']
    
    track_data = [id, acousticness, danceability, energy, key, mode, instrumentalness, liveness, loudness, speechiness, valence, tempo, time_signature]
    return track_data

```

```python
track_list = []
for i in range(len(track_ids)):
    track_data = getTrackFeatures(track_ids[i])
    track_list.append(track_data)

top200_playlist = pd.DataFrame(track_list, columns = ['ID','Acousticness', 'Danceablility', 'Energy', 'Key', 'Mode', 'Instrumentalness', 'Liveness', 'Loudness', 'Speechiness', 'Valence', 'Tempo', 'Time Signature'])
top200_playlist.head(10)
```

```python
top200_new = pd.concat([top200,top200_playlist], axis=1)
```
